```python
import customtkinter as ctk
from PIL import Image, ImageDraw
import time
import threading

# IMPORT LOGIN PAGE
#from login_update import SafePathAuthWindow

class SafePathAuthWindow(ctk.CTk):

    def __init__(self):

        super().__init__()

        self.geometry("1000x700")
        self.title("SafePath Portal")
        self.resizable(False, False)

        self.current_mode = "signin"

        # BACKGROUND
        self.bg_label = ctk.CTkLabel(self, text="")
        self.bg_label.place(x=0, y=0, relwidth=1, relheight=1)

        self.bind("<Configure>", self.render_background)

        # BUILD UI
        self.build_auth_card()
        
        self.bg_label.configure(image=gradient_img)
        self.bg_label.image = gradient_img

    
    def build_auth_card(self):

        self.card = ctk.CTkFrame(
            self,
            width=420,
            height=480,
            fg_color=COLOR_CARD_BG,
            corner_radius=24,
            border_width=1,
            border_color="#D9C2A3"
        )

        self.card.place(relx=0.5, rely=0.5, anchor="center")
        self.card.pack_propagate(False)

        # TITLE
        self.title_lbl = ctk.CTkLabel(
            self.card,
            text="Login",
            font=("Georgia", 32, "bold"),
            text_color=COLOR_SIDEBAR_DARK
        )

        self.title_lbl.pack(pady=(45, 30))


   
    def render_background(self, event=None):

        gradient_img = create_gradient_image(
            1000,
            700,
            COLOR_ACCENT_ORANGE,
            COLOR_SIDEBAR_DARK
        )

# ==========================================
# 🎨 COLOR SYSTEM
# ==========================================
COLOR_DARK = "#4B1E0F"
COLOR_ORANGE = "#B85C2E"
COLOR_LIGHT = "#F3EBDD"
COLOR_TEXT = "#FFFFFF"

ctk.set_appearance_mode("light")

# ==========================================
# 🖼️ CREATE GRADIENT BACKGROUND
# ==========================================
def create_gradient(width, height, color1, color2):

    base = Image.new("RGB", (width, height), color1)

    draw = ImageDraw.Draw(base)

    r1, g1, b1 = int(color1[1:3], 16), int(color1[3:5], 16), int(color1[5:7], 16)
    r2, g2, b2 = int(color2[1:3], 16), int(color2[3:5], 16), int(color2[5:7], 16)

    for y in range(height):

        blend = y / height

        r = int(r1 + (r2 - r1) * blend)
        g = int(g1 + (g2 - g1) * blend)
        b = int(b1 + (b2 - b1) * blend)

        draw.line([(0, y), (width, y)], fill=(r, g, b))

    return ctk.CTkImage(light_image=base, size=(width, height))

# ==========================================
# 🚀 SPLASH SCREEN CLASS
# ==========================================
class SplashScreen(ctk.CTk):

    def __init__(self):

        super().__init__()

        self.geometry("1000x700")
        self.overrideredirect(True)

        # CENTER WINDOW
        screen_width = self.winfo_screenwidth()
        screen_height = self.winfo_screenheight()

        x = int((screen_width / 2) - (1000 / 2))
        y = int((screen_height / 2) - (700 / 2))

        self.geometry(f"1000x700+{x}+{y}")

        # ==================================
        # BACKGROUND
        # ==================================
        bg_image = create_gradient(
            1000,
            700,
            COLOR_ORANGE,
            COLOR_DARK
        )

        bg_label = ctk.CTkLabel(
            self,
            image=bg_image,
            text=""
        )

        bg_label.place(
            relwidth=1,
            relheight=1
        )

        # PREVENT IMAGE GC
        bg_label.image = bg_image

        # ==================================
        # APP TITLE
        # ==================================
        title = ctk.CTkLabel(
            self,
            text="SafePath",
            font=("Georgia", 52, "bold"),
            text_color=COLOR_TEXT
        )

        title.pack(pady=(220, 10))

        # ==================================
        # TAGLINE
        # ==================================
        subtitle = ctk.CTkLabel(
            self,
            text="Awareness • Protection • Freedom",
            font=("Arial", 18),
            text_color="#F3EBDD"
        )

        subtitle.pack()

        # ==================================
        # LOADING TEXT
        # ==================================
        loading = ctk.CTkLabel(
            self,
            text="Loading secure environment...",
            font=("Arial", 14),
            text_color="#EADFC9"
        )

        loading.pack(pady=(40, 10))

        # ==================================
        # PROGRESS BAR
        # ==================================
        self.progress = ctk.CTkProgressBar(
            self,
            width=300,
            progress_color=COLOR_LIGHT,
            fg_color="#8A4F27"
        )

        self.progress.pack(pady=10)

        self.progress.set(0)

        # START LOADING THREAD
        threading.Thread(
            target=self.load_app
        ).start()

    # ======================================
    # LOAD APP
    # ======================================
    def load_app(self):

        for i in range(101):

            time.sleep(0.02)

            self.progress.set(i / 100)

        # CLOSE SPLASH
        self.destroy()

        # OPEN LOGIN
        login_window = SafePathAuthWindow()
        login_window.mainloop()

# ==========================================
# 🚀 RUN APP
# ==========================================
if __name__ == "__main__":

    app = SplashScreen()
    app.mainloop()
```

```python
"""
SafePath — Unified Production Build
Flow: Onboarding (Splash) ➔ Login / Sign Up Portal ➔ Workspace Panel Navigation
Integrated Modules: Dashboard, Situation Checker, Report Case, and Datastore Records
"""

import customtkinter as ctk
import psycopg2
import math
import csv
from tkinter import Canvas, messagebox, filedialog, ttk
from PIL import Image, ImageDraw

# ══════════════════════════════════════════════════════════════
#  SHARED STYLE & PALETTE DESIGN SYSTEM
# ══════════════════════════════════════════════════════════════
COLOR_SIDEBAR        = "#4B1E0F"   # Deep rich chocolate brown
COLOR_WORKSPACE_BG   = "#F3EBDD"   # Warm parchment cream 
COLOR_ACCENT_ORANGE  = "#B85C2E"   # Terracotta accent orange
COLOR_CARD_DARK      = "#5A2D0C"   # Deep brown status backing
COLOR_CARD_MID       = "#6B3F1D"   # Medium brown status surface
COLOR_CARD_LIGHT     = "#D9C2A3"   # Soft clay accent card
COLOR_CARD_BG        = "#EADFC9"   # Tan authorization backing
COLOR_TEXT_DARK      = "#3E2A1F"   # Primary text contrast charcoal
COLOR_TEXT_LIGHT     = "#C8A87A"   # Matte gold accent
COLOR_GRADIENT_TOP   = "#D4701A"   # Amber center glow
COLOR_GRADIENT_BOT   = "#2B1005"   # Near black boundary shade
COLOR_PROGRESS_TRACK = "#5A2808"   # Progress rail dark shadow
COLOR_PROGRESS_FILL  = "#F5E6D0"   # Linear loader cream fill

ctk.set_appearance_mode("light")

# ══════════════════════════════════════════════════════════════
#  RISK KEYWORD CONFIGURATION MATRIX
# ══════════════════════════════════════════════════════════════
RISK_KEYWORDS = {
    "cannot leave": 45, "not allowed to leave": 45, "leave": 10, "locked": 40, "restricted": 30,
    "passport": 45, "documents": 35, "id withheld": 35,
    "unpaid": 35, "without being paid": 40, "not paid": 35, "salary withheld": 35,
    "working for free": 45, "forced to work": 45, "worked for years": 25, "long hours": 20,
    "threat": 30, "threatening": 30, "threatened": 30, "fear": 15, "unsafe": 20, "afraid": 20,
    "forced": 35, "coercion": 40, "trafficking": 50, "false promise": 40, "recruited": 20,
    "transported": 25, "underage": 45, "child labour": 50, "isolated": 25, "monitored": 20,
    "not allowed contact": 35
}

# ══════════════════════════════════════════════════════════════
#  DATABASE ARCHITECTURE LAYER
# ══════════════════════════════════════════════════════════════
def db_connect():
    return psycopg2.connect(
        dbname="safepath",
        user="postgres",
        password="cos101",
        host="localhost",
        port="5432"
    )

def create_users_table():
    conn = db_connect()
    cur  = conn.cursor()
    cur.execute("""
        CREATE TABLE IF NOT EXISTS users (
            id       SERIAL PRIMARY KEY,
            username VARCHAR(100) UNIQUE NOT NULL,
            password VARCHAR(100) NOT NULL
        )
    """)
    cur.execute("""
        CREATE TABLE IF NOT EXISTS reports (
            id            SERIAL PRIMARY KEY,
            username      VARCHAR(100),
            location      VARCHAR(150),
            incident_type VARCHAR(150),
            urgency       VARCHAR(50),
            description   TEXT,
            anonymous     BOOLEAN DEFAULT FALSE
        )
    """)
    conn.commit()
    cur.close()
    conn.close()

def verify_or_register_user(username, password, mode="signin"):
    if not username or not password:
        return False, "Please fill in all fields."
    conn = db_connect()
    cur  = conn.cursor()
    try:
        cur.execute("SELECT password FROM users WHERE username = %s", (username,))
        row = cur.fetchone()
        if mode == "signin":
            if row is None: return False, "UserNotFound"
            return (True, "Login successful!") if row[0] == password else (False, "Incorrect password.")
        elif mode == "signup":
            if row is not None: return False, "Username already exists."
            cur.execute("INSERT INTO users (username, password) VALUES (%s, %s)", (username, password))
            conn.commit()
            return True, "Account created successfully!"
    except Exception as e:
        return False, str(e)
    finally:
        cur.close()
        conn.close()

def fetch_dashboard_metrics():
    try:
        conn = db_connect()
        cur = conn.cursor()
        cur.execute("SELECT COUNT(*) FROM reports")
        total_reports = cur.fetchone()[0]
        cur.execute("SELECT COUNT(*) FROM reports WHERE urgency ILIKE 'High%'")
        high_risk = cur.fetchone()[0]
        awareness, helped = total_reports + 5, total_reports
        cur.execute("SELECT incident_type, location FROM reports ORDER BY id DESC LIMIT 5")
        recent_reports = [f"{inc} reported in {loc}" for inc, loc in cur.fetchall()]
        cur.close()
        conn.close()
        return {"total": total_reports, "high_risk": high_risk, "awareness": awareness, "helped": helped, "recent": recent_reports}
    except Exception:
        return {"total": 0, "high_risk": 0, "awareness": 5, "helped": 0, "recent": []}

# ══════════════════════════════════════════════════════════════
#  GRAPHICAL UTILITIES
# ═══════════════════════════════════════════════════════════
def hex_to_rgb(hex_color):
    h = hex_color.lstrip("#")
    return tuple(int(h[i:i+2], 16) for i in (0, 2, 4))

def lerp_color(c1, c2, t):
    return tuple(int(c1[i] + (c2[i] - c1[i]) * t) for i in range(3))

def rgb_to_hex(rgb):
    return "#{:02x}{:02x}{:02x}".format(*rgb)

def create_linear_gradient_image(width, height, color_top, color_bot):
    img  = Image.new("RGB", (width, height))
    draw = ImageDraw.Draw(img)
    r1, g1, b1 = hex_to_rgb(color_top)
    r2, g2, b2 = hex_to_rgb(color_bot)
    for y in range(height):
        t = y / height
        draw.line([(0, y), (width, y)], fill=(
            int(r1 + (r2 - r1) * t), int(g1 + (g2 - g1) * t), int(b1 + (b2 - b1) * t)
        ))
    return ctk.CTkImage(light_image=img, size=(width, height))

# ══════════════════════════════════════════════════════════════
#  SCREEN 1 ➔ ONBOARDING SPLASH OVERVIEW
# ══════════════════════════════════════════════════════════════
class OnboardingScreen(ctk.CTk):
    W, H, STEPS, STEP_MS = 700, 480, 80, 32
    MESSAGES = ["Loading…", "Preparing your safe space…", "Almost ready…", "Welcome."]

    def __init__(self):
        super().__init__()
        self.title("SafePath")
        self.geometry(f"{self.W}x{self.H}")
        self.resizable(False, False)
        self.configure(fg_color=COLOR_GRADIENT_BOT)
        self._build_gradient_canvas()
        self._build_content()
        self.after(300, self._animate_progress)

    def _build_gradient_canvas(self):
        self._canvas = Canvas(self, width=self.W, height=self.H, bd=0, highlightthickness=0)
        self._canvas.place(x=0, y=0)
        cx, cy = self.W // 2, self.H // 2
        max_r = math.hypot(cx, cy)
        inner, mid, outer = hex_to_rgb(COLOR_GRADIENT_TOP), hex_to_rgb(COLOR_ACCENT_ORANGE), hex_to_rgb(COLOR_GRADIENT_BOT)
        for i in range(60, -1, -1):
            t = i / 60
            rx, ry = max_r * t * (self.W / max_r) * 0.72, max_r * t * (self.H / max_r) * 0.72
            color = lerp_color(inner, mid, min(t * 2, 1)) if t < 0.5 else lerp_color(mid, outer, (t - 0.5) * 2)
            self._canvas.create_oval(cx - rx, cy - ry, cx + rx, cy + ry, fill=rgb_to_hex(color), outline="")

    def _build_content(self):
        overlay = ctk.CTkFrame(self, fg_color="transparent")
        overlay.place(relx=0.5, rely=0.5, anchor="center")
        logo_ring = ctk.CTkFrame(overlay, width=90, height=90, fg_color=COLOR_SIDEBAR, corner_radius=45, border_color=COLOR_ACCENT_ORANGE, border_width=2)
        logo_ring.pack(pady=(0, 12))
        logo_ring.pack_propagate(False)
        ctk.CTkLabel(logo_ring, text="🌿", font=("Segoe UI Emoji", 36)).place(relx=0.5, rely=0.5, anchor="center")
        ctk.CTkLabel(overlay, text="SafePath", font=("Georgia", 32, "bold"), text_color=COLOR_WORKSPACE_BG).pack(pady=(0, 22))
        self._bar_track = ctk.CTkFrame(overlay, width=220, height=6, fg_color=COLOR_PROGRESS_TRACK, corner_radius=3)
        self._bar_track.pack(pady=(0, 10))
        self._bar_track.pack_propagate(False)
        self._bar_fill = ctk.CTkFrame(self._bar_track, width=0, height=6, fg_color=COLOR_PROGRESS_FILL, corner_radius=3)
        self._bar_fill.place(x=0, y=0)
        self._loading_lbl = ctk.CTkLabel(overlay, text=self.MESSAGES[0], font=("Segoe UI", 12, "italic"), text_color=COLOR_TEXT_LIGHT)
        self._loading_lbl.pack()

    def _animate_progress(self, step=0):
        if step > self.STEPS:
            self._loading_lbl.configure(text=self.MESSAGES[3])
            self.after(700, self._transition_to_login)
            return
        eased = 1 - (1 - (step / self.STEPS)) ** 2
        self._bar_fill.configure(width=max(int(220 * eased), 1))
        if step == 0: self._loading_lbl.configure(text=self.MESSAGES[0])
        elif step == 25: self._loading_lbl.configure(text=self.MESSAGES[1])
        elif step == 55: self._loading_lbl.configure(text=self.MESSAGES[2])
        self.after(self.STEP_MS, lambda: self._animate_progress(step + 1))

    def _transition_to_login(self, alpha=1.0):
        if alpha <= 0:
            self.destroy()
            LoginScreen().mainloop()
            return
        self.attributes("-alpha", alpha)
        self.after(18, lambda: self._transition_to_login(round(alpha - 0.05, 2)))

# ══════════════════════════════════════════════════════════════
#  SCREEN 2 ➔ SECURE ACCOUNT GATEWAY
# ══════════════════════════════════════════════════════════════
class LoginScreen(ctk.CTk):
    W, H = 1000, 700

    def __init__(self):
        super().__init__()
        self.title("SafePath Portal")
        self.geometry(f"{self.W}x{self.H}")
        self.resizable(False, False)
        self.current_mode = "signin"
        self._bg_label = ctk.CTkLabel(self, text="")
        self._bg_label.place(x=0, y=0, relwidth=1, relheight=1)
        self._render_background()
        self._build_auth_card()

    def _render_background(self):
        gradient = create_linear_gradient_image(self.W, self.H, COLOR_ACCENT_ORANGE, COLOR_SIDEBAR)
        self._bg_label.configure(image=gradient)
        self._bg_label.image = gradient   

    def _build_auth_card(self):
        self._card = ctk.CTkFrame(self, width=420, height=480, fg_color=COLOR_CARD_BG, corner_radius=24, border_width=1, border_color="#D9C2A3")
        self._card.place(relx=0.5, rely=0.5, anchor="center")
        self._card.pack_propagate(False)

        self._title_lbl = ctk.CTkLabel(self._card, text="Login", font=("Georgia", 32, "bold"), text_color=COLOR_SIDEBAR)
        self._title_lbl.pack(pady=(45, 30))
        self._user_entry = ctk.CTkEntry(self._card, placeholder_text="Username", width=320, height=45, corner_radius=10, fg_color=COLOR_WORKSPACE_BG, text_color=COLOR_TEXT_DARK, border_color="#D9C2A3")
        self._user_entry.pack(pady=12)
        self._pass_entry = ctk.CTkEntry(self._card, placeholder_text="Password", show="•", width=320, height=45, corner_radius=10, fg_color=COLOR_WORKSPACE_BG, text_color=COLOR_TEXT_DARK, border_color="#D9C2A3")
        self._pass_entry.pack(pady=12)

        self._action_btn = ctk.CTkButton(self._card, text="Login", command=self._execute_auth, fg_color=COLOR_SIDEBAR, hover_color="#33140A", text_color="white", width=320, height=48, corner_radius=12, font=("Georgia", 16, "bold"))
        self._action_btn.pack(pady=(25, 15))
        self._toggle_lbl = ctk.CTkLabel(self._card, text="Don't have an account? Sign up here", font=("Segoe UI", 12, "underline"), text_color=COLOR_TEXT_DARK, cursor="hand2")
        self._toggle_lbl.pack(pady=5)
        self._toggle_lbl.bind("<Button-1>", lambda e: self._switch_mode())

    def _switch_mode(self):
        self.current_mode = "signup" if self.current_mode == "signin" else "signin"
        if self.current_mode == "signin":
            self._title_lbl.configure(text="Login")
            self._action_btn.configure(text="Login", fg_color=COLOR_SIDEBAR)
            self._toggle_lbl.configure(text="Don't have an account? Sign up here")
        else:
            self._title_lbl.configure(text="Create Account")
            self._action_btn.configure(text="Register Account", fg_color=COLOR_ACCENT_ORANGE)
            self._toggle_lbl.configure(text="Already have an account? Sign in here")

    def _execute_auth(self):
        username = self._user_entry.get().strip()
        password = self._pass_entry.get().strip()
        success, message = verify_or_register_user(username, password, mode=self.current_mode)
        if success:
            messagebox.showinfo("SafePath", message)
            self.destroy()
            DashboardWindow(username=username).mainloop()
        else:
            if message == "UserNotFound":
                if messagebox.askyesno("User Not Found", f"No account found for '{username}'.\nCreate account?"): self._switch_mode()
            else: messagebox.showerror("Authentication Error", message)

# ══════════════════════════════════════════════════════════════
#  SCREEN 3 ➔ MAIN WORKSPACE PANEL
# ══════════════════════════════════════════════════════════════
class DashboardWindow(ctk.CTk):
    def __init__(self, username="User"):
        super().__init__()
        self.title("SafePath Operational Workspace")
        self.geometry("1440x900")
        self.username = username
        self.sidebar_buttons = {}
        self.current_view = ""  # Keeps tracking of loops to mitigate background UI update errors
        
        # LEFT NAVIGATION PANEL
        sidebar = ctk.CTkFrame(self, width=260, fg_color=COLOR_SIDEBAR, corner_radius=0)
        sidebar.pack(side="left", fill="y")
        sidebar.pack_propagate(False)

        logo = ctk.CTkLabel(sidebar, text="SafePath", font=("Georgia", 32, "bold"), text_color="white")
        logo.pack(pady=(40, 30), anchor="w", padx=30)

        menu_items = ["Dashboard", "Situation Check", "Report Case", "Records", "Insights", "Resources", "Echoes of Home", "Settings"]
        for item in menu_items:
            if item == "Dashboard": command = self.load_dashboard_view
            elif item == "Situation Check": command = self.load_situation_checker_view
            elif item == "Report Case": command = self.load_report_case_view
            elif item == "Records": command = self.load_records_view
            else: command = lambda item_name=item: print(f"{item_name} view is active.")

            btn = ctk.CTkButton(
                sidebar, text=item, command=command, fg_color="transparent",
                hover_color="#8A4F27", text_color="white", font=("Arial", 14), anchor="w", height=40
            )
            btn.pack(fill="x", padx=20, pady=5)
            self.sidebar_buttons[item] = btn

        tagline = ctk.CTkLabel(sidebar, text="Awareness. Protection. Freedom.", font=("Arial", 11, "italic"), text_color="#D9C2A3")
        tagline.pack(side="bottom", pady=25)

        # RIGHT WORKSPACE CONTAINER
        self.main_frame = ctk.CTkFrame(self, fg_color=COLOR_WORKSPACE_BG, corner_radius=0)
        self.main_frame.pack(side="right", fill="both", expand=True)

        self.load_dashboard_view()

    def clear_workspace(self):
        """Clears running workspace views safely and resets sidebar styles."""
        for widget in self.main_frame.winfo_children():
            widget.destroy()
        for item, btn in self.sidebar_buttons.items():
            btn.configure(fg_color="transparent", font=("Arial", 14, "normal"))

    # ══════════════════════════════════════════════════════════
    #  VIEW CONTROLLER 1: DASHBOARD HOME
    # ══════════════════════════════════════════════════════════
    def load_dashboard_view(self):
        self.clear_workspace()
        self.current_view = "Dashboard"
        self.sidebar_buttons["Dashboard"].configure(fg_color="#C47A3A", font=("Arial", 14, "bold"))

        title_lbl = ctk.CTkLabel(self.main_frame, text="Dashboard", font=("Georgia", 38, "bold"), text_color=COLOR_TEXT_DARK)
        title_lbl.pack(anchor="nw", padx=40, pady=(40, 5))

        subtitle_lbl = ctk.CTkLabel(self.main_frame, text=f"Welcome back, {self.username}.", font=("Arial", 16), text_color="#5E4B3C")
        subtitle_lbl.pack(anchor="nw", padx=40, pady=(0, 25))

        workspace_grid = ctk.CTkFrame(self.main_frame, fg_color="transparent")
        workspace_grid.pack(fill="both", expand=True, padx=40, pady=(0, 40))

        workspace_grid.columnconfigure(0, weight=1)
        workspace_grid.columnconfigure(1, weight=1)
        workspace_grid.rowconfigure(0, weight=1)
        workspace_grid.rowconfigure(1, weight=1)
        workspace_grid.rowconfigure(2, weight=2) 

        total_card = ctk.CTkFrame(workspace_grid, fg_color=COLOR_CARD_DARK, corner_radius=18)
        total_card.grid(row=0, column=0, sticky="nsew", padx=(0, 15), pady=10)
        ctk.CTkLabel(total_card, text="Total Reports", font=("Arial", 16), text_color="#F3EBDD").pack(anchor="nw", padx=25, pady=(25, 5))
        self.lbl_total_val = ctk.CTkLabel(total_card, text="--", font=("Georgia", 42, "bold"), text_color="white")
        self.lbl_total_val.pack(anchor="nw", padx=25)

        highrisk_card = ctk.CTkFrame(workspace_grid, fg_color=COLOR_ACCENT_ORANGE, corner_radius=18)
        highrisk_card.grid(row=0, column=1, sticky="nsew", padx=(15, 0), pady=10)
        ctk.CTkLabel(highrisk_card, text="High Risk Flagged", font=("Arial", 16, "bold"), text_color=COLOR_SIDEBAR).pack(anchor="nw", padx=25, pady=(25, 5))
        self.lbl_highrisk_val = ctk.CTkLabel(highrisk_card, text="--", font=("Georgia", 42, "bold"), text_color=COLOR_SIDEBAR)
        self.lbl_highrisk_val.pack(anchor="nw", padx=25)

        awareness_card = ctk.CTkFrame(workspace_grid, fg_color=COLOR_CARD_LIGHT, corner_radius=18)
        awareness_card.grid(row=1, column=0, sticky="nsew", padx=(0, 15), pady=10)
        ctk.CTkLabel(awareness_card, text="Awareness Check", font=("Arial", 16), text_color=COLOR_SIDEBAR).pack(anchor="nw", padx=25, pady=(25, 5))
        self.lbl_awareness_val = ctk.CTkLabel(awareness_card, text="--", font=("Georgia", 42, "bold"), text_color=COLOR_SIDEBAR)
        self.lbl_awareness_val.pack(anchor="nw", padx=25)

        helped_card = ctk.CTkFrame(workspace_grid, fg_color=COLOR_CARD_MID, corner_radius=18)
        helped_card.grid(row=1, column=1, sticky="nsew", padx=(15, 0), pady=10)
        ctk.CTkLabel(helped_card, text="People helped", font=("Arial", 16), text_color="#F3EBDD").pack(anchor="nw", padx=25, pady=(25, 5))
        self.lbl_helped_val = ctk.CTkLabel(helped_card, text="--", font=("Georgia", 42, "bold"), text_color="white")
        self.lbl_helped_val.pack(anchor="nw", padx=25)

        risk_assess_card = ctk.CTkFrame(workspace_grid, fg_color=COLOR_CARD_DARK, corner_radius=18)
        risk_assess_card.grid(row=2, column=0, sticky="nsew", padx=(0, 15), pady=(15, 0))
        ctk.CTkLabel(risk_assess_card, text="Unsure about a situation?", font=("Georgia", 26, "bold"), text_color="#F3EBDD").pack(anchor="center", pady=(35, 10))
        ctk.CTkLabel(risk_assess_card, text="Use our AI-driven analysis tool to identify signs\nof exploitation of forced labour.", font=("Arial", 15), text_color="#D9C2A3", justify="center").pack(anchor="center", pady=(0, 25))
        
        btn_assess = ctk.CTkButton(risk_assess_card, text="Start Risk Assessment", command=self.load_situation_checker_view, fg_color=COLOR_ACCENT_ORANGE, hover_color="#A44E25", text_color="white", font=("Arial", 15, "bold"), height=45, corner_radius=10)
        btn_assess.pack(anchor="center", ipadx=20)

        recent_reports_card = ctk.CTkFrame(workspace_grid, fg_color="#EADFC9", corner_radius=18, border_width=1, border_color="#D9C2A3")
        recent_reports_card.grid(row=2, column=1, sticky="nsew", padx=(15, 0), pady=(15, 0))
        ctk.CTkLabel(recent_reports_card, text="Recent Reports", font=("Georgia", 22, "bold"), text_color=COLOR_TEXT_DARK).pack(anchor="nw", padx=25, pady=(25, 10))
        self.feed_container = ctk.CTkFrame(recent_reports_card, fg_color="transparent")
        self.feed_container.pack(fill="both", expand=True, padx=25)

        self.update_dashboard_view()

    def update_dashboard_view(self):
        if self.current_view != "Dashboard" or not hasattr(self, 'lbl_total_val'): 
            return
        data = fetch_dashboard_metrics()
        self.lbl_total_val.configure(text=data["total"])
        self.lbl_highrisk_val.configure(text=data["high_risk"])
        self.lbl_awareness_val.configure(text=data["awareness"])
        self.lbl_helped_val.configure(text=data["helped"])

        for widget in self.feed_container.winfo_children(): widget.destroy()
        if data["recent"]:
            for report_text in data["recent"]:
                item_lbl = ctk.CTkLabel(self.feed_container, text=f"• {report_text}", font=("Arial", 14), text_color=COLOR_TEXT_DARK, anchor="w")
                item_lbl.pack(fill="x", pady=6)
                ctk.CTkFrame(self.feed_container, height=1, fg_color="#D9C2A3").pack(fill="x", pady=(2, 6))
        else:
            ctk.CTkLabel(self.feed_container, text="No new notifications found.", font=("Arial", 14, "italic"), text_color="#766555").pack(pady=20)
        
        self.after(5000, self.update_dashboard_view)

    # ══════════════════════════════════════════════════════════
    #  VIEW CONTROLLER 2: SITUATION CHECKER
    # ══════════════════════════════════════════════════════════
    def load_situation_checker_view(self):
        self.clear_workspace()
        self.current_view = "SituationCheck"
        self.sidebar_buttons["Situation Check"].configure(fg_color="#C47A3A", font=("Arial", 14, "bold"))

        title_lbl = ctk.CTkLabel(self.main_frame, text="Situation Checker", font=("Georgia", 38, "bold"), text_color=COLOR_TEXT_DARK)
        title_lbl.pack(anchor="nw", padx=40, pady=(40, 5))

        subtitle_lbl = ctk.CTkLabel(self.main_frame, text="Describe your circumstances below to cross-evaluate indicators safely.", font=("Arial", 15), text_color="#5E4B3C")
        subtitle_lbl.pack(anchor="nw", padx=40, pady=(0, 25))

        checker_split = ctk.CTkFrame(self.main_frame, fg_color="transparent")
        checker_split.pack(fill="both", expand=True, padx=40, pady=(0, 40))
        checker_split.columnconfigure(0, weight=3)
        checker_split.columnconfigure(1, weight=2)
        checker_split.rowconfigure(0, weight=1)

        left_input_panel = ctk.CTkFrame(checker_split, fg_color="#F7F0E4", corner_radius=18, border_width=1, border_color="#8A4F27")
        left_input_panel.grid(row=0, column=0, sticky="nsew", padx=(0, 15))
        
        ctk.CTkLabel(left_input_panel, text="What is your current work or living situation?", font=("Georgia", 18, "bold"), text_color=COLOR_TEXT_DARK).pack(anchor="w", padx=25, pady=(25, 10))
        
        self.situation_textbox = ctk.CTkTextbox(left_input_panel, fg_color="white", text_color=COLOR_TEXT_DARK, border_color="#8A4F27", border_width=1, font=("Arial", 14), corner_radius=12)
        self.situation_textbox.pack(fill="both", expand=True, padx=25, pady=(0, 20))

        btn_analyze = ctk.CTkButton(left_input_panel, text="Analyze Risk Indicators", command=self.execute_risk_analysis, fg_color="#8A4F27", hover_color="#5A2D0C", text_color="white", font=("Arial", 14, "bold"), height=42, corner_radius=10)
        btn_analyze.pack(fill="x", padx=25, pady=(0, 25))

        right_results_panel = ctk.CTkFrame(checker_split, fg_color=COLOR_CARD_DARK, corner_radius=18)
        right_results_panel.grid(row=0, column=1, sticky="nsew", padx=(15, 0))

        ctk.CTkLabel(right_results_panel, text="Evaluation Matrix", font=("Georgia", 20, "bold"), text_color="white").pack(anchor="w", padx=25, pady=(25, 20))
        
        self.risk_result_label = ctk.CTkLabel(right_results_panel, text="NOT ANALYZED", font=("Georgia", 32, "bold"), text_color="#D9C2A3")
        self.risk_result_label.pack(pady=20)

        legend_box = ctk.CTkFrame(right_results_panel, fg_color="transparent")
        legend_box.pack(anchor="w", padx=25, pady=10)
        ctk.CTkLabel(legend_box, text="🔴 High Risk (70+ points)\n🟠 Caution (35+ points)\n🟢 Low Risk (0+ points)", font=("Arial", 13), text_color="#F3EBDD", justify="left").pack()

        self.indicators_label = ctk.CTkLabel(right_results_panel, text="No indicators running yet.\nEnter text and press analyze to calculate threats.", justify="left", font=("Arial", 14), text_color="#F3EBDD", wraplength=320)
        self.indicators_label.pack(anchor="nw", padx=25, pady=(30, 20))

    def execute_risk_analysis(self):
        text = self.situation_textbox.get("1.0", "end").lower()
        score = 0
        found_indicators = []

        for keyword, points in RISK_KEYWORDS.items():
            if keyword in text:
                score += points
                found_indicators.append(keyword)

        if score >= 70:
            risk_level, color = "HIGH RISK", "#FF4C4C"
        elif score >= 35:
            risk_level, color = "CAUTION", "#FF9F43"
        else:
            risk_level, color = "LOW RISK", "#2ED573"

        self.risk_result_label.configure(text=risk_level, text_color=color)

        if found_indicators:
            self.indicators_label.configure(
                text=f"Score Weight: {score}\n\nIndicators Found:\n• " + "\n• ".join(found_indicators)
            )
        else:
            self.indicators_label.configure(text=f"Score Weight: {score}\n\nNo major flags or modern exploitation indicators detected.")

    # ══════════════════════════════════════════════════════════
    #  VIEW CONTROLLER 3: REPORT CASE INLINE FORM
    # ══════════════════════════════════════════════════════════
    def load_report_case_view(self):
        self.clear_workspace()
        self.current_view = "ReportCase"
        self.sidebar_buttons["Report Case"].configure(fg_color="#C47A3A", font=("Arial", 14, "bold"))

        title_lbl = ctk.CTkLabel(self.main_frame, text="Report Case Portal", font=("Georgia", 38, "bold"), text_color=COLOR_TEXT_DARK)
        title_lbl.pack(anchor="nw", padx=40, pady=(40, 5))

        subtitle_lbl = ctk.CTkLabel(self.main_frame, text="Provide case information securely. Data is protected by strict protocols.", font=("Arial", 15), text_color="#5E4B3C")
        subtitle_lbl.pack(anchor="nw", padx=40, pady=(0, 25))

        form_card = ctk.CTkFrame(self.main_frame, fg_color="#EADFC9", corner_radius=18, border_width=1, border_color="#D9C2A3")
        form_card.pack(fill="both", expand=True, padx=40, pady=(0, 40))

        form_card.columnconfigure(0, weight=1)
        form_card.columnconfigure(1, weight=1)

        ctk.CTkLabel(form_card, text="Incident Location", font=("Georgia", 15, "bold"), text_color=COLOR_TEXT_DARK).grid(row=0, column=0, padx=30, pady=(25, 5), sticky="w")
        self.entry_location = ctk.CTkEntry(form_card, placeholder_text="City, State or Country", height=40, fg_color=COLOR_WORKSPACE_BG, text_color=COLOR_TEXT_DARK, border_color="#D9C2A3")
        self.entry_location.grid(row=1, column=0, padx=30, pady=(0, 15), sticky="ew")

        ctk.CTkLabel(form_card, text="Incident Type", font=("Georgia", 15, "bold"), text_color=COLOR_TEXT_DARK).grid(row=0, column=1, padx=30, pady=(25, 5), sticky="w")
        self.opt_incident = ctk.CTkOptionMenu(form_card, values=["Forced Labour", "Human Trafficking", "Withheld Documents", "Unpaid Salaries", "Other Exploitation"], height=40, fg_color="#8A4F27", button_color="#5A2D0C", button_hover_color="#33140A")
        self.opt_incident.grid(row=1, column=1, padx=30, pady=(0, 15), sticky="ew")

        ctk.CTkLabel(form_card, text="Urgency Level", font=("Georgia", 15, "bold"), text_color=COLOR_TEXT_DARK).grid(row=2, column=0, padx=30, pady=(10, 5), sticky="w")
        self.opt_urgency = ctk.CTkOptionMenu(form_card, values=["Low Urgency", "Medium Urgency", "High (Immediate Action Required)"], height=40, fg_color="#8A4F27", button_color="#5A2D0C", button_hover_color="#33140A")
        self.opt_urgency.grid(row=3, column=0, padx=30, pady=(0, 15), sticky="ew")

        ctk.CTkLabel(form_card, text="Privacy Preference", font=("Georgia", 15, "bold"), text_color=COLOR_TEXT_DARK).grid(row=2, column=1, padx=30, pady=(10, 5), sticky="w")
        self.var_anonymous = ctk.BooleanVar(value=False)
        self.chk_anonymous = ctk.CTkCheckBox(form_frame:=form_card, text="Submit report anonymously", variable=self.var_anonymous, font=("Arial", 14), text_color=COLOR_TEXT_DARK, checkbox_width=22, checkbox_height=22, fg_color="#8A4F27", hover_color="#5A2D0C")
        self.chk_anonymous.grid(row=3, column=1, padx=30, pady=(5, 15), sticky="w")

        ctk.CTkLabel(form_card, text="Detailed Case Description", font=("Georgia", 15, "bold"), text_color=COLOR_TEXT_DARK).grid(row=4, column=0, columnspan=2, padx=30, pady=(10, 5), sticky="w")
        self.txt_description = ctk.CTkTextbox(form_card, fg_color="white", text_color=COLOR_TEXT_DARK, border_color="#D9C2A3", border_width=1, font=("Arial", 14), corner_radius=10)
        self.txt_description.grid(row=5, column=0, columnspan=2, padx=30, pady=(0, 20), sticky="nsew")
        form_card.rowconfigure(5, weight=1) 

        self.btn_submit_report = ctk.CTkButton(form_card, text="Submit Confidential Report", command=self.execute_report_submission, fg_color="#5A2D0C", hover_color="#33140A", text_color="white", font=("Georgia", 16, "bold"), height=48, corner_radius=12)
        self.btn_submit_report.grid(row=6, column=0, columnspan=2, padx=30, pady=(0, 25), sticky="ew")

    def execute_report_submission(self):
        location = self.entry_location.get().strip()
        incident_type = self.opt_incident.get()
        urgency = self.opt_urgency.get()
        anonymous = self.var_anonymous.get()
        description = self.txt_description.get("1.0", "end").strip()

        if not location or not description:
            messagebox.showerror("Validation Error", "Please provide a valid incident location and a comprehensive case description.")
            return

        reported_by = "Anonymous" if anonymous else self.username

        try:
            conn = db_connect()
            cur = conn.cursor()
            cur.execute("""
                INSERT INTO reports (username, location, incident_type, urgency, description, anonymous)
                VALUES (%s, %s, %s, %s, %s, %s)
            """, (reported_by, location, incident_type, urgency, description, anonymous))
            conn.commit()
            cur.close()
            conn.close()

            messagebox.showinfo("Success", "Your report has been received securely.\nOperational units have been notified.")
            self.load_dashboard_view()
        except Exception as e:
            messagebox.showerror("Database Execution Error", f"Failed to submit case data safely:\n{str(e)}")

    # ══════════════════════════════════════════════════════════
    #  VIEW CONTROLLER 4: HISTORIC RECORDS DATAGRID
    # ══════════════════════════════════════════════════════════
    def load_records_view(self):
        self.clear_workspace()
        self.current_view = "Records"
        self.sidebar_buttons["Records"].configure(fg_color="#C47A3A", font=("Arial", 14, "bold"))

        title_lbl = ctk.CTkLabel(self.main_frame, text="Database Records", font=("Georgia", 32, "bold"), text_color=COLOR_TEXT_DARK)
        title_lbl.pack(anchor="nw", padx=40, pady=(30, 5))

        subtitle_lbl = ctk.CTkLabel(self.main_frame, text="Review secure historic entries and download operational data logs.", font=("Arial", 15), text_color="#5E4B3C")
        subtitle_lbl.pack(anchor="nw", padx=40, pady=(0, 20))

        # Search Bar and Export Panel
        search_frame = ctk.CTkFrame(self.main_frame, fg_color="transparent")
        search_frame.pack(fill="x", padx=40, pady=(10, 0))

        self.search_entry = ctk.CTkEntry(search_frame, width=320, placeholder_text="Search by User, Location, or Incident...")
        self.search_entry.pack(side="left", padx=(0, 10))

        search_btn = ctk.CTkButton(search_frame, text="Search", fg_color=COLOR_CARD_DARK, hover_color="#7A431A", text_color="white", width=100, command=lambda: self.fetch_and_render_reports(self.search_entry.get()))
        search_btn.pack(side="left", padx=5)

        clear_btn = ctk.CTkButton(search_frame, text="Clear", fg_color="transparent", hover_color=COLOR_CARD_LIGHT, text_color=COLOR_TEXT_DARK, width=80, border_width=1, border_color=COLOR_CARD_DARK, command=self.clear_search_query)
        clear_btn.pack(side="left", padx=5)

        export_btn = ctk.CTkButton(search_frame, text="Export as CSV", fg_color=COLOR_ACCENT_ORANGE, hover_color="#A04810", text_color="white", command=self.export_records_to_csv)
        export_btn.pack(side="right")

        # Table Canvas View Layout
        table_card = ctk.CTkFrame(self.main_frame, fg_color="#F7F0E4", corner_radius=18, border_width=1, border_color="#D9C2A3")
        table_card.pack(fill="both", expand=True, padx=40, pady=25)

        style = ttk.Style()
        style.theme_use("default")
        style.configure("Treeview", background="#F3E7D7", foreground=COLOR_TEXT_DARK, fieldbackground="#F3E7D7", rowheight=38, bordercolor="#6B3E26", borderwidth=1, font=("Arial", 11))
        style.configure("Treeview.Heading", background=COLOR_CARD_DARK, foreground="white", font=("Georgia", 12, "bold"))
        style.map("Treeview", background=[("selected", COLOR_ACCENT_ORANGE)])

        columns = ("ID", "Username", "Location", "Incident Type", "Urgency")
        self.records_tree = ttk.Treeview(table_card, columns=columns, show="headings", height=15)

        for col in columns:
            self.records_tree.heading(col, text=col)
            if col in ["ID", "Urgency"]:
                self.records_tree.column(col, width=100, anchor="center")
            else:
                self.records_tree.column(col, width=220, anchor="w")

        self.records_tree.pack(fill="both", expand=True, padx=20, pady=20)
        self.fetch_and_render_reports()

    def fetch_and_render_reports(self, search_text=""):
        try:
            conn = db_connect()
            cur = conn.cursor()
            query = """
                SELECT id, username, location, incident_type, urgency
                FROM reports
                WHERE username ILIKE %s OR location ILIKE %s OR incident_type ILIKE %s
                ORDER BY id ASC
            """
            wildcard_val = f"%{search_text}%"
            cur.execute(query, (wildcard_val, wildcard_val, wildcard_val))
            rows = cur.fetchall()

            for row in self.records_tree.get_children():
                self.records_tree.delete(row)

            for row in rows:
                self.records_tree.insert("", "end", values=row)

            cur.close()
            conn.close()
        except Exception as e:
            messagebox.showerror("Query Failure", f"Failed to sync logs securely from database:\n{str(e)}")

    def clear_search_query(self):
        self.search_entry.delete(0, 'end')
        self.fetch_and_render_reports("")

    def export_records_to_csv(self):
        try:
            conn = db_connect()
            cur = conn.cursor()
            cur.execute("SELECT id, username, location, incident_type, urgency FROM reports ORDER BY id ASC")
            rows = cur.fetchall()
            cur.close()
            conn.close()

            if not rows:
                messagebox.showwarning("Export Void", "There are currently no structural dashboard records to export.")
                return

            file_path = filedialog.asksaveasfilename(defaultextension=".csv", filetypes=[("CSV Files", "*.csv")], title="Save Operational Log Ledger")
            if file_path:
                with open(file_path, mode="w", newline="", encoding="utf-8") as file:
                    writer = csv.writer(file)
                    writer.writerow(["ID", "Username", "Location", "Incident Type", "Urgency"])
                    writer.writerows(rows)
                messagebox.showinfo("Export Successful", f"Database record archive successfully exported to:\n{file_path}")
        except Exception as e:
            messagebox.showerror("Export Error", f"Failed to dump dataset data arrays onto file storage:\n{str(e)}")

# ══════════════════════════════════════════════════════════════
#  RUNTIME EXECUTION ENGINE
# ═══════════════════════════════════════════════════════════
if __name__ == "__main__":
    create_users_table()  
    OnboardingScreen().mainloop()
```

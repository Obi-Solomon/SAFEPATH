```python
import customtkinter as ctk
import psycopg2
from tkinter import messagebox

ctk.set_appearance_mode("light")

# DATABASE CONNECTION
def connect():
    return psycopg2.connect(
        dbname="safepath",
        user="postgres",
        password="cos101",
        host="localhost",
        port="5432"
    )

# SUBMIT FUNCTION
def submit_report():
    try:
        conn = connect()
        cur = conn.cursor()

        cur.execute("""
            INSERT INTO reports (
                username,
                location,
                incident_type,
                urgency,
                description,
                anonymous
            )
            VALUES (%s, %s, %s, %s, %s, %s)
        """, (
            username_entry.get(),
            location_entry.get(),
            incident_entry.get(),
            urgency_entry.get(),
            description_box.get("1.0", "end"),
            anonymous_var.get()
        ))

        conn.commit()
        cur.close()
        conn.close()

        messagebox.showinfo("Success", "Report submitted successfully")

    except Exception as e:
        messagebox.showerror("Database Error", str(e))

# MAIN WINDOW
root = ctk.CTk()
root.geometry("1200x700")
root.title("SafePath")

# COLORS
sidebar_color = "#4B1E0F"
main_bg = "#F3EBDD"
button_color = "#C47A3A"

# SIDEBAR
sidebar = ctk.CTkFrame(
    root,
    width=220,
    fg_color=sidebar_color,
    corner_radius=0
)

sidebar.pack(side="left", fill="y")

logo = ctk.CTkLabel(
    sidebar,
    text="SafePath",
    font=("Georgia", 28, "bold"),
    text_color="white"
)

logo.pack(pady=40)

menu_items = [
    "Dashboard",
    "Situation Check",
    "Report Case",
    "Records",
    "Insights",
    "Resources",
    "Echoes of Home",
    "Settings"
]

for item in menu_items:
    btn = ctk.CTkButton(
        sidebar,
        text=item,
        fg_color=button_color if item == "Report Case" else "transparent",
        hover_color="#8A4F27",
        anchor="w"
    )
    btn.pack(fill="x", padx=20, pady=8)

# MAIN CONTENT
main_frame = ctk.CTkFrame(
    root,
    fg_color=main_bg,
    corner_radius=0
)

main_frame.pack(side="right", fill="both", expand=True)

title = ctk.CTkLabel(
    main_frame,
    text="Report Case",
    font=("Georgia", 32, "bold"),
    text_color="#3E2A1F"
)

title.pack(anchor="nw", padx=40, pady=(30,10))

subtitle = ctk.CTkLabel(
    main_frame,
    text="Official incident intake. Please ensure all mandatory fields are completed.",
    font=("Arial", 14),
    text_color="#5E4B3C"
)

subtitle.pack(anchor="nw", padx=40)

# FORM FRAME
form_frame = ctk.CTkFrame(
    main_frame,
    fg_color="#F7F0E4",
    border_width=2,
    border_color="#6B3E26"
)

form_frame.pack(padx=40, pady=30)

# ROW 1
username_entry = ctk.CTkEntry(form_frame, width=250, placeholder_text="Username")
username_entry.grid(row=0, column=0, padx=20, pady=20)

location_entry = ctk.CTkEntry(form_frame, width=250, placeholder_text="Location")
location_entry.grid(row=0, column=1, padx=20, pady=20)

# ROW 2
urgency_entry = ctk.CTkEntry(form_frame, width=250, placeholder_text="Urgency Level")
urgency_entry.grid(row=1, column=0, padx=20, pady=20)

incident_entry = ctk.CTkEntry(form_frame, width=250, placeholder_text="Incident Type")
incident_entry.grid(row=1, column=1, padx=20, pady=20)

# DESCRIPTION
description_box = ctk.CTkTextbox(
    form_frame,
    width=550,
    height=150
)

description_box.grid(row=2, column=0, columnspan=2, padx=20, pady=20)

# ANONYMOUS CHECKBOX
anonymous_var = ctk.BooleanVar()

anonymous_check = ctk.CTkCheckBox(
    form_frame,
    text="Submit report anonymously",
    variable=anonymous_var
)

anonymous_check.grid(row=3, column=0, sticky="w", padx=20)

# SUBMIT BUTTON
submit_btn = ctk.CTkButton(
    form_frame,
    text="Submit Report",
    width=250,
    height=45,
    fg_color=button_color,
    hover_color="#8A4F27",
    command=submit_report
)

submit_btn.grid(row=4, column=0, columnspan=2, pady=30)

root.mainloop()
```

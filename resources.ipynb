```python
import customtkinter as ctk
from PIL import Image, ImageDraw
import tkinter as tk


# COLOR SCHEMA TOKENS (Strictly synchronized with settings.ipynb)

SIDEBAR_BG      = "#3B1A0A"  # Deep luxury dark brown
SIDEBAR_ACTIVE  = "#7B3A10"  # Vibrant amber active background
SIDEBAR_TEXT    = "#F5E6D0"  # Warm cream text
CONTENT_BG      = "#F0E0C8"  # Smooth sand background canvas
CARD_BG         = "#E8D0A8"  # Slightly darker card surface
CARD_BORDER     = "#C8A87A"  # Structural outline stroke
ACCENT_BTN      = "#C06020"  # Signature action orange
ACCENT_BTN_HOV  = "#A04810"  # Hover state variation
TEXT_DARK       = "#2A1005"  # Primary typography brown
TEXT_MID        = "#6B4020"  # Subtitle reading gray/brown
TEXT_LIGHT      = "#9B7050"  # Muted tagline details

# Force light mode workspace style matching setup defaults
ctk.set_appearance_mode("light")


#  VECTOR CORE GENERATOR (Figma Emblem Reconstruction)

def generate_vector_logo_image():
    """Generates a high-resolution version of the golden fist/broken chain brand motif."""
    size = (160, 160)
    img = Image.new("RGBA", size, (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)
    
    center_x, center_y = 80, 80
    outer_ring_radius = 55
    
    # Outer Ring Container
    draw.ellipse(
        [(center_x - outer_ring_radius, center_y - outer_ring_radius),
         (center_x + outer_ring_radius, center_y + outer_ring_radius)],
        outline="#D4A373", width=3
    )
    
    # Broken Side Chain Links
    draw.arc([(center_x - 43, center_y - 15), (center_x - 20, center_y + 15)], start=40, end=320, fill="#A67C52", width=4)
    draw.arc([(center_x + 20, center_y - 15), (center_x + 43, center_y + 15)], start=220, end=140, fill="#A67C52", width=4)
    
    # Upward Solid Fist
    draw.rectangle([(center_x - 14, center_y + 10), (center_x + 14, center_y + 40)], fill="#D4A373")
    draw.rounded_rectangle([(center_x - 18, center_y - 20), (center_x + 18, center_y + 10)], radius=5, fill="#C29468")
    draw.polygon([(center_x - 18, center_y - 2), (center_x - 4, center_y + 8), (center_x - 18, center_y + 12)], fill="#D4A373")
    
    for offset_x in [-9, 0, 9]:
        draw.line([(center_x + offset_x, center_y - 20), (center_x + offset_x, center_y - 8)], fill=SIDEBAR_BG, width=2)
        
    return ctk.CTkImage(light_image=img, dark_image=img, size=(32, 32))



#  MAIN UNIFIED APPLICATION RUNTIME

class SafePathResourcesWindow(ctk.CTk):
    def __init__(self):
        super().__init__()

        self.title("SafePath — Resources & Support")
        self.geometry("1050x720")
        self.minimum_width = 980
        self.minimum_height = 650
        self.resizable(True, True)
        self.configure(fg_color=CONTENT_BG)

        
        # NAVIGATION SIDEBAR FRAMEWORK
        self.sidebar = ctk.CTkFrame(self, width=240, fg_color=SIDEBAR_BG, corner_radius=0)
        self.sidebar.pack(side="left", fill="y")
        self.sidebar.pack_propagate(False)

        # Logo row Header
        self.logo_frame = ctk.CTkFrame(self.sidebar, fg_color="transparent")
        self.logo_frame.pack(fill="x", padx=18, pady=(24, 10))

        self.logo_icon = ctk.CTkLabel(self.logo_frame, text="", fg_color="transparent")
        self.logo_icon.pack(side="left")
        self.logo_icon.configure(image=generate_vector_logo_image())

        self.logo_label = ctk.CTkLabel(
            self.logo_frame, text=" SafePath",
            font=("Georgia", 20, "bold"), text_color=SIDEBAR_TEXT
        )
        self.logo_label.pack(side="left", padx=4)

        self.sep = ctk.CTkFrame(self.sidebar, height=1, fg_color="#5A2E14")
        self.sep.pack(fill="x", padx=14, pady=(6, 16))

        # Sidebar Menu Items Configuration Array
        NAV_ITEMS = [
            ("🏠", "Dashboard",       False),
            ("⚠️", "Situation Check",  False),
            ("📋", "Report Case",      False),
            ("📁", "Records",          False),
            ("📊", "Insights",         False),
            ("🎧", "Resources",        True),   # Active State Indicator Set
            ("🌐", "Echoes of Home",   False),
            ("⚙️", "Settings",         False),
        ]

        for icon, label, active in NAV_ITEMS:
            self.make_nav_item(self.sidebar, icon, label, active)

        # Tagline Anchor Bottom Block
        self.tagline = ctk.CTkLabel(
            self.sidebar,
            text="Awareness. Protection. Freedom",
            font=("Georgia", 10, "italic"),
            text_color=TEXT_LIGHT,
            wraplength=200,
            justify="center"
        )
        self.tagline.pack(side="bottom", pady=22)

        
        #  MAIN SCROLLABLE CONTENT VIEWPORT
        
        self.content_scroll = ctk.CTkScrollableFrame(
            self, fg_color=CONTENT_BG,
            scrollbar_button_color=CARD_BORDER,
            scrollbar_button_hover_color=ACCENT_BTN
        )
        self.content_scroll.pack(side="left", fill="both", expand=True)

        # Header Titles Area
        ctk.CTkLabel(
            self.content_scroll, text="Resources & Support",
            font=("Georgia", 28, "bold"), text_color=TEXT_DARK, anchor="w"
        ).pack(anchor="w", padx=36, pady=(32, 2))

        ctk.CTkLabel(
            self.content_scroll,
            text="Access immediate emergency assistance, local support services and helplines.",
            font=("Segoe UI", 13), text_color=TEXT_MID, anchor="w"
        ).pack(anchor="w", padx=36, pady=(0, 24))

       
        # CARD 1: EMERGENCY & HELPLINE ROW BLOCK
       
        self.emergency_card = ctk.CTkFrame(
            self.content_scroll, fg_color=CARD_BG, corner_radius=14,
            border_color=CARD_BORDER, border_width=1
        )
        self.emergency_card.pack(fill="x", padx=36, pady=(0, 24))

        ctk.CTkLabel(
            self.emergency_card, text="EMERGENCY & HELPLINE",
            font=("Segoe UI", 13, "bold"), text_color=TEXT_DARK, anchor="w"
        ).pack(anchor="w", padx=20, pady=(16, 12))

        # Inner Container with soft padding borders
        self.inner_call_box = ctk.CTkFrame(self.emergency_card, fg_color="#E0C498", corner_radius=10)
        self.inner_call_box.pack(fill="x", padx=20, pady=(0, 20))

        self.helpline_text_lbl = ctk.CTkLabel(
            self.inner_call_box,
            text="National Anti-Trafficking Helpline:\n07030000203  OR  080 CALL NAPTIP (08002255627847)",
            font=("Segoe UI", 14, "bold"), text_color=TEXT_DARK, justify="center"
        )
        self.helpline_text_lbl.pack(pady=(18, 14))

        # Native Functional Action Trigger Button
        self.call_btn = ctk.CTkButton(
            self.inner_call_box,
            text="📞   [ CALL NOW ]",
            font=("Segoe UI", 14, "bold"),
            fg_color=ACCENT_BTN,
            hover_color=ACCENT_BTN_HOV,
            text_color="#FFFFFF",
            corner_radius=8,
            height=42,
            width=240,
            command=self.execute_emergency_dial
        )
        self.call_btn.pack(pady=(0, 18))

        
        #  CARD 2: STRUCTURAL SUPPORT DIRECTORY TABLE
        
        self.directory_card = ctk.CTkFrame(
            self.content_scroll, fg_color=CARD_BG, corner_radius=14,
            border_color=CARD_BORDER, border_width=1
        )
        self.directory_card.pack(fill="x", padx=36, pady=(0, 32))

        # Custom grid layout container simulating design matrix
        self.table_grid = ctk.CTkFrame(self.directory_card, fg_color="transparent")
        self.table_grid.pack(fill="both", expand=True, padx=16, pady=16)
        self.table_grid.columnconfigure(0, weight=4, uniform="group1")
        self.table_grid.columnconfigure(1, weight=5, uniform="group1")

        # --- Matrix Column Headers ---
        self.th1 = ctk.CTkLabel(self.table_grid, text="NGOs", font=("Segoe UI", 14, "bold"), text_color=TEXT_DARK, anchor="center")
        self.th1.grid(row=0, column=0, sticky="ew", pady=(0, 8))
        
        self.th2 = ctk.CTkLabel(self.table_grid, text="Contact Details", font=("Segoe UI", 14, "bold"), text_color=TEXT_DARK, anchor="center")
        self.th2.grid(row=0, column=1, sticky="ew", pady=(0, 8))

        # --- Divider Rule ---
        self.grid_sep = ctk.CTkFrame(self.table_grid, height=2, fg_color=TEXT_DARK)
        self.grid_sep.grid(row=1, column=0, columnspan=2, sticky="ew", pady=(0, 14))

        # --- Directory Database Row Elements ---
        NGO_DATASET = [
            {
                "name": "Devatop Centre for African\nDevelopment",
                "details": "Kings Park Estate IV, Kukwaba District by\nGaladimawa Roundabout, Abuja, Nigeria.\n+234 903 000 2362"
            },
            {
                "name": "Women's Consortium of Nigeria\n(WOCON)",
                "details": "House 229A, Ikoyi Crescent, Dolphin Estate,\nIkoyi, Lagos.\n+234 803 719 0133, +234 816 659 6954"
            },
            {
                "name": "National Agency for the Prohibition\nof Trafficking in Persons (NAPTIP)",
                "details": "No. 2028, Dalaba Street, Wuse Zone 5, FCT\nAbuja\n07030000203\ninfo@naptip.gov.ng"
            }
        ]

        current_row = 2
        for idx, entry in enumerate(NGO_DATASET):
            # Left cell (NGO Name descriptor)
            cell_left = ctk.CTkLabel(
                self.table_grid, text=entry["name"],
                font=("Segoe UI", 13, "bold"), text_color=TEXT_DARK,
                justify="center", anchor="center"
            )
            cell_left.grid(row=current_row, column=0, sticky="nsew", padx=8, pady=16)

            # Right cell (Contact endpoints block)
            cell_right = ctk.CTkLabel(
                self.table_grid, text=entry["details"],
                font=("Segoe UI", 12), text_color=TEXT_DARK,
                justify="center", anchor="center"
            )
            cell_right.grid(row=current_row, column=1, sticky="nsew", padx=8, pady=16)

            # Add separator rules between rows, skipping the last element block
            if idx < len(NGO_DATASET) - 1:
                row_sep = ctk.CTkFrame(self.table_grid, height=1, fg_color=CARD_BORDER)
                row_sep.grid(row=current_row + 1, column=0, columnspan=2, sticky="ew", pady=4)
                current_row += 2
            else:
                current_row += 1

    def make_nav_item(self, parent, icon, label, active):
        """Builds standardized isolated navigation list tabs."""
        bg = SIDEBAR_ACTIVE if active else "transparent"
        radius = 8 if active else 0

        row = ctk.CTkFrame(parent, fg_color=bg, corner_radius=radius, height=40)
        row.pack(fill="x", padx=12, pady=3)
        row.pack_propagate(False)

        lbl = ctk.CTkLabel(
            row, text=f"{icon}   {label}",
            font=("Segoe UI", 13, "bold" if active else "normal"),
            text_color=SIDEBAR_TEXT, anchor="w"
        )
        lbl.pack(side="left", padx=16, pady=6)
        return row

    def execute_emergency_dial(self):
        """Simulates native calling integration routines."""
        tk.messagebox.showinfo(
            "SafePath Protocol", 
            "Emergency Request Initialized!\nDialing NAPTIP Headquarters Response Line: 07030000203..."
        )


#  JUPYTER KERNEL SAFE-RUN COMMAND

# This blocks Jupyter from freezing if a window instance is already active
try:
    if 'app' in globals() and app.winfo_exists():
        app.destroy()
except Exception:
    pass

app = SafePathResourcesWindow()
app.mainloop()
```

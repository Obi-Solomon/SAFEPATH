```python
import customtkinter as ctk


SIDEBAR_BG      = "#3B1A0A"  
SIDEBAR_ACTIVE  = "#7B3A10"   
SIDEBAR_TEXT    = "#F5E6D0"   
CONTENT_BG      = "#F0E0C8"   
CARD_BG         = "#E8D0A8"   
CARD_BORDER     = "#C8A87A"  
ACCENT_BTN      = "#C06020"   
ACCENT_BTN_HOV  = "#A04810"   
INPUT_BG        = "#F5EAD8"   
TEXT_DARK       = "#2A1005"  
TEXT_MID        = "#6B4020"   
TEXT_LIGHT      = "#9B7050"   


ctk.set_appearance_mode("light")
ctk.set_default_color_theme("blue")   

app = ctk.CTk()
app.title("Safepath — Settings")
app.geometry("950x680")
app.resizable(True, True)
app.configure(fg_color=CONTENT_BG)



sidebar = ctk.CTkFrame(app, width=220, fg_color=SIDEBAR_BG, corner_radius=0)
sidebar.pack(side="left", fill="y")
sidebar.pack_propagate(False)

# — Logo row —
logo_frame = ctk.CTkFrame(sidebar, fg_color="transparent")
logo_frame.pack(fill="x", padx=18, pady=(22, 10))

logo_icon = ctk.CTkLabel(
    logo_frame, text="🌿", font=("Segoe UI Emoji", 24),
    text_color=SIDEBAR_TEXT
)
logo_icon.pack(side="left")

logo_label = ctk.CTkLabel(
    logo_frame, text="  Safepath",
    font=("Georgia", 18, "bold"), text_color=SIDEBAR_TEXT
)
logo_label.pack(side="left")


sep = ctk.CTkFrame(sidebar, height=1, fg_color=SIDEBAR_ACTIVE)
sep.pack(fill="x", padx=12, pady=(4, 14))


NAV_ITEMS = [
    ("🏠", "Dashboard",       False),
    ("⚠️", "Situation Check",  False),
    ("📋", "Report Case",      False),
    ("📁", "Records",          False),
    ("📊", "Insights",         False),
    ("🔗", "Resources",        False),
    ("🌐", "Echoes of Home",   False),
    ("⚙️", "Settings",         True),  
]

nav_buttons = []

def make_nav_item(parent, icon, label, active):
    """Creates a single navigation row."""
    bg     = SIDEBAR_ACTIVE if active else "transparent"
    radius = 8 if active else 0

    row = ctk.CTkFrame(parent, fg_color=bg, corner_radius=radius, height=38)
    row.pack(fill="x", padx=10, pady=2)
    row.pack_propagate(False)

    ctk.CTkLabel(
        row, text=f"{icon}  {label}",
        font=("Segoe UI", 13, "bold" if active else "normal"),
        text_color=SIDEBAR_TEXT, anchor="w"
    ).pack(side="left", padx=14, pady=6)

    return row

for icon, label, active in NAV_ITEMS:
    make_nav_item(sidebar, icon, label, active)

# — Tagline at bottom —
tagline = ctk.CTkLabel(
    sidebar,
    text="Awareness. Protection. Freedom",
    font=("Georgia", 9, "italic"),
    text_color=TEXT_LIGHT,
    wraplength=180,
    justify="center"
)
tagline.pack(side="bottom", pady=18)



content_scroll = ctk.CTkScrollableFrame(
    app, fg_color=CONTENT_BG,
    scrollbar_button_color=CARD_BORDER,
    scrollbar_button_hover_color=ACCENT_BTN
)
content_scroll.pack(side="left", fill="both", expand=True)

content = content_scroll   


ctk.CTkLabel(
    content, text="Settings",
    font=("Georgia", 26, "bold"), text_color=TEXT_DARK, anchor="w"
).pack(anchor="w", padx=32, pady=(28, 2))

ctk.CTkLabel(
    content,
    text="Manage your account, privacy, notification and preferences.",
    font=("Segoe UI", 12), text_color=TEXT_MID, anchor="w"
).pack(anchor="w", padx=32, pady=(0, 20))



def make_card(parent, title, icon, padx=32, pady=(0, 16)):
    """Returns the inner frame of a labelled card."""
    card = ctk.CTkFrame(
        parent, fg_color=CARD_BG,
        corner_radius=12,
        border_color=CARD_BORDER, border_width=1
    )
    card.pack(fill="x", padx=padx, pady=pady)

    header = ctk.CTkFrame(card, fg_color="transparent")
    header.pack(fill="x", padx=16, pady=(14, 6))

    ctk.CTkLabel(
        header, text=f"{icon}  {title}",
        font=("Georgia", 14, "bold"), text_color=TEXT_DARK
    ).pack(side="left")

    return card



row1 = ctk.CTkFrame(content, fg_color="transparent")
row1.pack(fill="x", padx=32, pady=(0, 16))
row1.columnconfigure(0, weight=1)
row1.columnconfigure(1, weight=1)

acct_card = ctk.CTkFrame(
    row1, fg_color=CARD_BG, corner_radius=12,
    border_color=CARD_BORDER, border_width=1
)
acct_card.grid(row=0, column=0, sticky="nsew", padx=(0, 8))

ctk.CTkLabel(
    acct_card, text="👤  Account Settings",
    font=("Georgia", 14, "bold"), text_color=TEXT_DARK, anchor="w"
).pack(anchor="w", padx=16, pady=(14, 8))

username_entry = ctk.CTkEntry(
    acct_card, placeholder_text="Username",
    fg_color=INPUT_BG, border_color=CARD_BORDER,
    text_color=TEXT_DARK, placeholder_text_color=TEXT_LIGHT,
    height=36, font=("Segoe UI", 12)
)
username_entry.pack(fill="x", padx=16, pady=(0, 8))

password_entry = ctk.CTkEntry(
    acct_card, placeholder_text="Password",
    show="•",
    fg_color=INPUT_BG, border_color=CARD_BORDER,
    text_color=TEXT_DARK, placeholder_text_color=TEXT_LIGHT,
    height=36, font=("Segoe UI", 12)
)
password_entry.pack(fill="x", padx=16, pady=(0, 16))


priv_card = ctk.CTkFrame(
    row1, fg_color=CARD_BG, corner_radius=12,
    border_color=CARD_BORDER, border_width=1
)
priv_card.grid(row=0, column=1, sticky="nsew", padx=(8, 0))

ctk.CTkLabel(
    priv_card, text="🛡  Privacy & Safety",
    font=("Georgia", 14, "bold"), text_color=TEXT_DARK, anchor="w"
).pack(anchor="w", padx=16, pady=(14, 12))

def privacy_toggle_row(parent, label, sublabel):
    """One labelled toggle row inside a card."""
    row = ctk.CTkFrame(parent, fg_color="transparent")
    row.pack(fill="x", padx=16, pady=(0, 12))

    text_col = ctk.CTkFrame(row, fg_color="transparent")
    text_col.pack(side="left", fill="x", expand=True)

    ctk.CTkLabel(
        text_col, text=label,
        font=("Segoe UI", 12, "bold"), text_color=TEXT_DARK, anchor="w"
    ).pack(anchor="w")
    ctk.CTkLabel(
        text_col, text=sublabel,
        font=("Segoe UI", 10), text_color=TEXT_LIGHT, anchor="w"
    ).pack(anchor="w")

    toggle = ctk.CTkSwitch(
        row, text="",
        fg_color=CARD_BORDER, progress_color=ACCENT_BTN,
        button_color=SIDEBAR_BG, button_hover_color=SIDEBAR_ACTIVE,
        width=46, height=22
    )
    toggle.pack(side="right", padx=4)

privacy_toggle_row(priv_card, "Anonymous Reports",  "Send report anonymously by default")
privacy_toggle_row(priv_card, "Hide Identity",      "Hide your identity in shared data")



notif_card = ctk.CTkFrame(
    content, fg_color=CARD_BG, corner_radius=12,
    border_color=CARD_BORDER, border_width=1
)
notif_card.pack(fill="x", padx=32, pady=(0, 16))

ctk.CTkLabel(
    notif_card, text="🔔  Notifications",
    font=("Georgia", 14, "bold"), text_color=TEXT_DARK, anchor="w"
).pack(anchor="w", padx=16, pady=(14, 10))

def notif_toggle_row(parent, label, sublabel):
    row = ctk.CTkFrame(parent, fg_color="transparent")
    row.pack(fill="x", padx=16, pady=(0, 10))

    toggle = ctk.CTkSwitch(
        row, text="",
        fg_color=CARD_BORDER, progress_color=ACCENT_BTN,
        button_color=SIDEBAR_BG, button_hover_color=SIDEBAR_ACTIVE,
        width=46, height=22
    )
    toggle.pack(side="left", padx=(0, 14))

    text_col = ctk.CTkFrame(row, fg_color="transparent")
    text_col.pack(side="left")

    ctk.CTkLabel(
        text_col, text=label,
        font=("Segoe UI", 12, "bold"), text_color=TEXT_DARK, anchor="w"
    ).pack(anchor="w")
    ctk.CTkLabel(
        text_col, text=sublabel,
        font=("Segoe UI", 10), text_color=TEXT_LIGHT, anchor="w"
    ).pack(anchor="w")

notif_toggle_row(notif_card, "Safety Alert",     "Receive important safety alert")
notif_toggle_row(notif_card, "Awareness update", "Receive updates and awareness tips.")

ctk.CTkFrame(notif_card, height=6, fg_color="transparent").pack()  # bottom padding



ctk.CTkLabel(
    content, text="🌐  Region & Language",
    font=("Georgia", 15, "bold"), text_color=TEXT_DARK, anchor="w"
).pack(anchor="w", padx=32, pady=(4, 8))


region_card = ctk.CTkFrame(
    content, fg_color=CARD_BG, corner_radius=12,
    border_color=CARD_BORDER, border_width=1
)
region_card.pack(fill="x", padx=32, pady=(0, 24))


region_inner = ctk.CTkFrame(region_card, fg_color="transparent")
region_inner.pack(fill="x", padx=16, pady=16)
region_inner.columnconfigure(0, weight=1)
region_inner.columnconfigure(1, weight=1)


left_col = ctk.CTkFrame(region_inner, fg_color="transparent")
left_col.grid(row=0, column=0, sticky="ew", padx=(0, 8))

ctk.CTkLabel(
    left_col, text="Region/Country",
    font=("Segoe UI", 11), text_color=TEXT_MID, anchor="w"
).pack(anchor="w", pady=(0, 4))

region_entry = ctk.CTkEntry(
    left_col,
    placeholder_text="📍  Nigeria",
    fg_color=INPUT_BG, border_color=CARD_BORDER,
    text_color=TEXT_DARK, placeholder_text_color=TEXT_DARK,
    height=38, font=("Segoe UI", 12)
)
region_entry.pack(fill="x")
region_entry.insert(0, "Nigeria")

right_col = ctk.CTkFrame(region_inner, fg_color="transparent")
right_col.grid(row=0, column=1, sticky="ew", padx=(8, 0))

ctk.CTkLabel(
    right_col, text="Language",
    font=("Segoe UI", 11), text_color=TEXT_MID, anchor="w"
).pack(anchor="w", pady=(0, 4))

lang_menu = ctk.CTkOptionMenu(
    right_col,
    values=["English"],
    fg_color=INPUT_BG, button_color=CARD_BORDER,
    button_hover_color=ACCENT_BTN, text_color=TEXT_DARK,
    dropdown_fg_color=INPUT_BG, dropdown_text_color=TEXT_DARK,
    dropdown_hover_color=CARD_BORDER,
    font=("Segoe UI", 12), height=38
)
lang_menu.set("English")
lang_menu.pack(fill="x")



def on_save():
    btn_save.configure(text="✓  Saved!", fg_color="#4A8C30")
    app.after(2000, lambda: btn_save.configure(
        text="💾  Save Changes", fg_color=ACCENT_BTN
    ))

btn_save = ctk.CTkButton(
    content,
    text="💾  Save Changes",
    font=("Georgia", 14, "bold"),
    fg_color=ACCENT_BTN,
    hover_color=ACCENT_BTN_HOV,
    text_color="#FFFFFF",
    corner_radius=10,
    height=46,
    command=on_save
)
btn_save.pack(fill="x", padx=32, pady=(0, 32))



app.mainloop()
```
